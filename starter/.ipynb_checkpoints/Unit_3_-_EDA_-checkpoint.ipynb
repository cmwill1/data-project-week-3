```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

#### Fill in blanks with a brief description of each plot type:

  
- Histogram: 
> [A histogram is a statistical graph that uses vertical bars to display quantitative data. The heights of the bars indicate the frequencies or relative frequencies of values in our data.]
- Density Plot / KDE
> [A density plot/KDE is a method for estimating the probability density function of a random variable in a data set.  ]
- Bar Chart
> [A bar chart shows values represented as vertical bars and can be shown in different colors to distinguish data. It can be used to show trends in data, and the comparison of multiple data sets side by side.]
- Line Plot
> [A Line plot is as a graph that shows data as points or marks above a number line, showing the frequency of each value in the data. ]
- Univariate scatter plot
> [The scatter diagram is a graph that pairs numerical data with one variable on each axis. This helps to show the relationship between these pairs and also can show when there is no correlation.]
- Violin Plot
> [A violin plot displays data peaks and it is a mix of a box plot and a kernel density plot. It shows how numerical data is distributed. Violin plots show summary statistics and the density of each variable.]
- Box Plot
> [A box plot is a graphical view of statistical data based on quartiles to show how a data set is distributed. (the minimum, first quartile, median, third quartile, and maximum.)]
- Pie Chart
> [ A pie chart is a graph in which a circle is divided into sections that each represent a proportion of the whole.]
- Rug Plot
> [A rug plot is a visualization of data for a single quantitative variable that is displayed as marks along an axis. It is used to visualise the distribution of the data.]

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn

import seaborn as sns
# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
plt.figure(figsize=(12,6))
ax = sns.histplot(data=df, x='age', palette="viridis", bins=8)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)
plt.title("Distribution of Manager Ages in Survey", fontsize=15)
```

##### Insights:

[As age increases, so does number of managers, there is a positive correlation. This remains true until age 65 where the amount of managers stays roughly the same.  ]

### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='industry',palette="Set2")

plt.xticks(rotation=90)
plt.xlabel("Industry", fontsize=12)
plt.ylabel("Number of Managers in Industry", fontsize=12)
plt.title("Industries in Survey", fontsize=15)
```

##### Insights:

[The industry type and number of managers is nearly equal across industries.]

### Plot for 'salary' column

```python
# create plot for salary column
sns.displot(data = df, x = 'salary', kind = 'hist', bins=40,)

plt.xlabel("Salary", fontsize=12)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=14)
```

```python
sns.displot(data = df, x = 'salary', kde = True, palette="Greens", edgecolor="black", bins=25, line_kws={"linewidth":2})

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=16);
```

```python
plt.figure(figsize=(12,6))

sns.violinplot(x=df['salary'])

plt.xlabel("Salary", fontsize=14)
plt.title("Density of Mana`ger Salaries");
```

##### Insights:

[The median salary among managers in this dataset is around 100,000. ]

### Plot for 'industry_experience' column

```python
# create plot for industry experience column
plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='industry_experience', bins=20)

plt.xlabel("Industry Experience ", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)
plt.title("Industry Experience Among Managers", fontsize=16);
```

##### Insights:

[A significant amount of managers in this data set have between 15-25 years of industry experience.]

### Plot for 'education' column

```python
# create plot for education levels

plt.figure(figsize=(10,10))

data = df['education'].value_counts()

labels = ["Bachelor's Degree", "Master's Degree", "Some College", "High School"]

colors = sns.color_palette('Accent')

plt.title("Manager's Education Levels", fontsize=16)

ax = plt.pie(data, colors=colors, labels=labels, autopct='%.0f%%')
```

##### Insights:

[The managers in this data-set hold a bacherlors degree, over other education types. ]

### Plot for 'gender' column

```python
# create plot for gender column
plt.figure(figsize=(10,10))

data = df['gender'].value_counts()

labels = ['Male', 'Female']

colors = sns.color_palette('rainbow')

ax = plt.pie(data, colors=colors, labels=labels, autopct='%.0f%%');
```

##### Insights:

[The distribution of male and female managers is equal in this data set.]

- age **standard deviation**
> The standard deviation is [12.758662] 

###### Reflection: How did mean and median differ above? Why do you think that is?

[The mean and median differed very slightly. The mean was 46.2 and the median is 47 which indicates there are no extreme outliers in the data set in this column.]
